```python
import numpy as np

def extract_features(arr):
    if type(arr) is not type(np.array([])):
      return [0] * 7
    # else
    return [
        np.mean(arr),
        np.std(arr),
        np.min(arr),
        np.max(arr),
        np.median(arr),
        np.percentile(arr, 25),  # First quartile (Q1)
        np.percentile(arr, 75)   # Third quartile (Q3)
    ]
```

```python
featureExtractedMurzymes = list()
featureExtractedNonMurzymes = list()

for i in range(len(murzymesReconstructed)):
    sample_features = []
    sample_features.extend(extract_features(murzymesReconstructed[i][0]))
    sample_features.extend(extract_features(murzymesReconstructed[i][1]))
    sample_features.extend(extract_features(murzymesReconstructed[i][2]))
    sample_features.extend(extract_features(murzymesReconstructed[i][3]))
    featureExtractedMurzymes.append(sample_features)

for j in range(len(nonMurzymesReconstructed)):
    sample_features = []
    sample_features.extend(extract_features(nonMurzymesReconstructed[j][0]))
    sample_features.extend(extract_features(nonMurzymesReconstructed[j][1]))
    sample_features.extend(extract_features(nonMurzymesReconstructed[j][2]))
    sample_features.extend(extract_features(nonMurzymesReconstructed[j][3]))
    featureExtractedNonMurzymes.append(sample_features)
```

## Scaling and Training the model
## LogisticRegression

```python
Samples, target = totalSamplesDf.iloc[:, :28], totalSamplesDf["target"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(Samples, target, test_size=0.2, random_state=16)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train_scaled, y_train)

log_y_pred = logreg.predict(X_test_scaled)
```

```python
print(list(log_y_pred), list(y_test.array), sep="\n")
```

# Using SVM - RBF Kernel

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix



svm_classifier = SVC(kernel='rbf', C=1.0, gamma='scale')  # RBF kernel is commonly used
svm_classifier.fit(X_train_scaled, y_train)

svm_y_pred = svm_classifier.predict(X_test_scaled)
```

```python
print(list(svm_y_pred), list(y_test.array), sep="\n")
```

```python
# Calculate accuracy
print(f"LogisticRegression Accuracy: {accuracy_score(y_test, log_y_pred)}")
print(f"SVM Accuracy: {accuracy_score(y_test, svm_y_pred)}")

print("\n","="*100,"\n")

# Classification report
print("LogisticRegression Report:\n")
print(classification_report(y_test, log_y_pred))
print("\n",'-'*100,"\n")
print("SVM Report:\n")
print(classification_report(y_test, svm_y_pred))

print("\n","="*100,"\n")

# Confusion matrix
print("LogisticRegression Confusion Matrix:")
print(confusion_matrix(y_test, log_y_pred))
print("SVM Confusion Matrix:")
print(confusion_matrix(y_test, svm_y_pred))
```

## Random Forest Ensemble

```python
from sklearn.ensemble import RandomForestClassifier  # For classification tasks

# Define hyperparameters (adjust as needed)
n_estimators = 100  # Number of trees in the forest
max_depth = 5  # Maximum depth of individual trees
min_samples_split = 2  # Minimum samples required to split a node
min_samples_leaf = 1  # Minimum samples required at each leaf node

# Create the random forest classifier object
model = RandomForestClassifier(n_estimators=n_estimators,
                              max_depth=max_depth,
                              min_samples_split=min_samples_split,
                              min_samples_leaf=min_samples_leaf)

# Train the model on the training data
model.fit(X_train_scaled, y_train)

# Make predictions on the testing data
rf_y_pred = model.predict(X_test)
```

## Feature Preparation

```python
with open("/content/murzymes.pkl", 'rb') as murzymes, open("/content/nonmurzymes.pkl", 'rb') as nonmurzymes:
  murzymesDict = pk.load(murzymes)
  nonMurzymesDict = pk.load(nonmurzymes)
```

```python
murzyme_keys = list(murzymesDict.keys())
non_murzyme_keys = list(nonMurzymesDict.keys())

feature1_murzymes = [list(v[0]) for k, v in murzymesDict.items()]
feature1_non_murzymes = [list(v[0]) for k, v in nonMurzymesDict.items()]

feature2_murzymes = [list(v[1]) for k, v in murzymesDict.items()]
feature2_non_murzymes = [list(v[1]) for k, v in nonMurzymesDict.items()]

feature3_murzymes = [list(v[2]) for k, v in murzymesDict.items()]
feature3_non_murzymes = [list(v[2]) for k, v in nonMurzymesDict.items()]

feature4_murzymes = [([0]*50 if type(v[3]) != np.ndarray else list(v[3])) for k, v in murzymesDict.items()]
feature4_non_murzymes = [([0]*50 if type(v[3]) != np.ndarray else list(v[3])) for k, v in nonMurzymesDict.items()]
```

```python
from sklearn.preprocessing import StandardScaler

scaler1 = StandardScaler()
scaler1.fit(feature1_murzymes+feature1_non_murzymes)

scaler2 = StandardScaler()
scaler2.fit(feature2_murzymes+feature2_non_murzymes)

scaler3 = StandardScaler()
scaler3.fit(feature3_murzymes+feature3_non_murzymes)

scaler4 = StandardScaler()
scaler4.fit(feature4_murzymes+feature4_non_murzymes)
```

```python
feature1_murzymes_scaled = scaler1.transform(feature1_murzymes)
feature1_non_murzymes_scaled = scaler1.transform(feature1_non_murzymes)

feature2_murzymes_scaled = scaler2.transform(feature2_murzymes)
feature2_non_murzymes_scaled = scaler2.transform(feature2_non_murzymes)

feature3_murzymes_scaled = scaler3.transform(feature3_murzymes)
feature3_non_murzymes_scaled = scaler3.transform(feature3_non_murzymes)

feature4_murzymes_scaled = scaler4.transform(feature4_murzymes)
feature4_non_murzymes_scaled = scaler4.transform(feature4_non_murzymes)
```
